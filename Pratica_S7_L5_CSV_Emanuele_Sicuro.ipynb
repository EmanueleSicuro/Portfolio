```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

dataset = pd.read_csv("dataset_climatico.csv")

# Valori anomali o errati
dataset.isnull().sum()

# Valori mancanti
dataset = dataset.dropna()


# Normalizzazione
columns_to_normalize = ['temperatura_media', 'precipitazioni', 'umidita', 'velocita_vento']

# Normalizzazione Z-score
scaler = StandardScaler()
dataset[columns_to_normalize] = scaler.fit_transform(dataset[columns_to_normalize])

dataset.to_csv("dataset_normalizzato.csv", index=False)
```

```python
# Calcolo le statistiche descrittive
desc_stats = dataset.describe()

# Media
media = dataset.mean()

# Mediana
mediana = dataset.median()

# Deviazione standard
deviazione_standard = dataset.std()

print("Statistiche descrittive:")
print(desc_stats)

print("\nMedia:")
print(media)

print("\nMediana:")
print(mediana)

print("\nDeviazione Standard:")
print(deviazione_standard)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dataset_normalizzato = pd.read_csv("dataset_normalizzato.csv")

colonne_da_visualizzare = ['temperatura_media', 'precipitazioni', 'umidita', 'velocita_vento']

# Istogrammi
for column in colonne_da_visualizzare:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=dataset, x=column, kde=True)
    plt.title(f"Istogramma di {column}")
    plt.xlabel(column)
    plt.ylabel("Frequenza")
    plt.grid(True)
    plt.show()
```

```python
# Box plots
for column in colonne_da_visualizzare:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=dataset, y=column)
    plt.title(f"Box plot di {column}")
    plt.ylabel(column)
    plt.grid(True)
    plt.show()
```

```python
correlazione = dataset_normalizzato.corr()

# Crea la heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlazione, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap della correlazione tra variabili meteorologiche")
plt.show()
```

In questo report, sono stati analizzati i dati climatici raccolti da diverse stazioni meteorologiche. Il dataset include le seguenti variabili: data_osservazione, temperatura_media, precipitazioni, umidita e velocita_vento.

**Pulizia dei Dati**:
Prima di procedere con l'analisi, i dati sono stati puliti per rimuovere eventuali valori mancanti o errati. Le righe contenenti valori mancanti sono state eliminate.

**Normalizzazione dei Dati**:
Le variabili temperatura_media, precipitazioni, umidit e velocit_vento sono state normalizzate utilizzando la tecnica della Z-score, al fine di standardizzarle e facilitare la comparazione tra di loro.

**Statistiche Descrittive**:
Sono state calcolate le statistiche descrittive (media, mediana e deviazione standard) per ogni variabile nel dataset. Queste statistiche forniscono una panoramica generale delle distribuzioni delle variabili.

**Analisi delle Correlazioni**:
Una heatmap  stata utilizzata per visualizzare la correlazione tra le variabili meteorologiche. Sono state identificate correlazioni significative tra temperatura_media e umidit, cos come tra temperatura_media e velocit_vento. In particolare,  stata osservata una correlazione negativa moderata tra temperatura_media e umidit, il che suggerisce che quando la temperatura aumenta, l'umidit tende a diminuire. D'altra parte,  stata osservata una correlazione positiva debole tra temperatura_media e velocit_vento, indicando una tendenza al aumento della velocit del vento con l'aumentare della temperatura.

Conclusioni:
L'analisi dei dati climatici ha fornito una visione approfondita delle relazioni tra le diverse variabili meteorologiche. Le correlazioni identificate possono essere utili per comprendere meglio i pattern climatici e prevedere eventuali cambiamenti nel tempo.
